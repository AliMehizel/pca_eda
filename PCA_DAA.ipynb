###  Application analyse en composante principale

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#data
data = pd.read_excel('tp_data.xlsx', sheet_name="4")

data_tab=pd.DataFrame(data)
data_tab
```

```python
#la matrice
mat_x = np.array(data_tab.values[:,2:],dtype=float)
#shape
mat_x.shape
```

### Mtrice des données

```python
mat_x
```

```python
from IPython.display import display, Math

def print_matrix(array):
    matrix = ''
    for row in array:
        try:
            for number in row:
                matrix += f'{number}&'
        except TypeError:
            matrix += f'{row}&'
        matrix = matrix[:-1] + r'\\'
    display(Math(r'\begin{bmatrix}'+matrix+r'\end{bmatrix}'))
print_matrix(mat_x)
```

```python
#description
data_tab.describe()
```

### Centre de gravité

```python
#centrage et réduction
#centre de gravité
g = np.mean(mat_x,0)
g
```

```python
print_matrix(g)
```

```python
#variance 
sigma_sq = np.var(mat_x,0)

print_matrix(sigma_sq)
```

```python
#ecart-type
sigma_ = np.std(mat_x,axis=0)
print_matrix(sigma_)
```

## Centrage et reduction

```python
#matrice centre reduit
x_norme = (mat_x - g) / sigma_sq
pd.DataFrame(x_norme)
```

```python
print_matrix(x_norme)
```

```python
#matrice de corrélation
corr_mat = (1/24)*np.matmul(x_norme.T, x_norme)
pd.DataFrame(corr_mat)
```

```python
print_matrix(corr_mat)
```

## Inertie et axes principale

```python
#valeur propre et vecteur propre
va_p, vec_prop = np.linalg.eig(corr_mat)
va_p
```

```python
print_matrix(vec_prop)
```

```python
#vecteur propre num 1
print_matrix(vec_prop[:,0])
```

```python
inertie_tt = np.sum(va_p)
inertie_tt
```

```python
#qte d'infos 1er axes principale
(va_p[0]*100)/inertie_tt
```

```python
#verifiaction
t = 0
for i in range(6):
    t = t+ (va_p[i]*100)/inertie_tt
    print(t)
```

## Axes normé

```python
axes_1 = vec_prop[:,0]
axes_2 = vec_prop[:,1]
#vecteur norme
print(np.linalg.norm(axes_1))
print(np.linalg.norm(axes_2))
```

## Componsantes principale

```python
#les corredonnées
corrd_1 = np.matmul(x_norme , axes_1)
corrd_2 = np.matmul(x_norme , axes_2)

print_matrix(corrd_1)
print_matrix(corrd_2)
```

```python
#new_corr = pd.DataFrame(np.vstack([corrd_1,corrd_2]).T)
new_corr = pd.DataFrame({'axes_1':corrd_1,'axes_2':corrd_2})
new_corr
```

```python
plt.scatter(x=new_corr['axes_1'],
                      y=new_corr['axes_2'],
                      c='DarkBlue',alpha=1)
ax =plt.gca()
ax.spines['bottom'].set_color('red')
ax.spines['top'].set_color('red')
ax.spines['left'].set_color('blue')
ax.spines['right'].set_color('blue')
```
